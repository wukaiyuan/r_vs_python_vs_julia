Make pairwise scatterplots
-------------------------

```julia
using Gadfly, Compose
function scatterplot_matrix(df::AbstractDataFrame, cols::Vector{Symbol})
    n = length(cols)
    cs = Array(Context, (n, n))
    for (i, coli) in enumerate(cols), (j, colj) in enumerate(cols)
        if i==j
            cs[i, j] = compose(context(), text(0.5, 0.5, string(coli), hcenter, vcenter), linewidth(0.01mm), 
            stroke(colorant"black"), fontsize(5))
        else
            cs[i, j] = render(plot(x=df[coli], y=df[colj], Guide.xlabel(nothing), Guide.ylabel(nothing), 
                Guide.yticks(label=false), Guide.xticks(label=false)))
        end
    end
    draw(SVGJS((5*n)cm, (5*n)cm), gridstack(cs))
end

scatterplot_matrix(nba, [:ast, :fg, :trb])
```

Make clusters of the players
--------------------------

```julia
using Clustering
badcolumns = Symbol[]
for nbacolumn in names(nba)
    if (size(findin(isna(nba[nbacolumn]),true),1) > 0) || (typeof(nba[nbacolumn]) == DataArrays.DataArray{UTF8String,1})
        push!(badcolumns,nbacolumn)
    end
end
nbaclean = nba[:, filter(x -> !(x in badcolumns), names(nba))]
matrix = transpose(convert(Array,nbaclean))
floatmatrix = convert(Array{Float64,2},matrix)
clusters = kmeans(floatmatrix, 5)
labels = assignments(clusters)
```

Plot players by cluster
----------------------

```julia
using MultivariateStats, Gadfly
M = fit(PCA, floatmatrix; maxoutdim=100)
plot_columns = transform(M,floatmatrix)[1:2,:]
plot_columns = transpose(plot_columns)
pl = plot(x=plot_columns[:,1], y=plot_columns[:,2], color=labels, Geom.point)
```

Splitting into training and test sets
-----------------------------------

```julia
n = size(nba,1)
is_train = shuffle([1:n] .> floor(n * .2))
train,test = nba[is_train,:],nba[!is_train,:]
```

Univariate linear regression
---------------------------

```julia
using GLM
data = DataFrame(X=train[:fg],Y=train[:ast])
OLS = glm(Y~X,data,Normal(),IdentityLink())
testdata = convert(Array{Float64,1},test[:fg])
model = coef(OLS)
predictions = model[2].*testdata + model[1]
```

Fit a random forest model
------------------------

```julia
using DecisionTree
predictor_columns = [:age, :mp, :fg, :trb, :stl, :blk]
yTrain = convert(Array{Float64,1},train[:ast])
xTrain = convert(Array{Float64,2},convert(Array,train[predictor_columns]))
model = build_forest(yTrain, xTrain, 2, 100, 1.0)
xTest = convert(Array{Float64,2},convert(Array,test[predictor_columns]))
predTest = apply_forest(model, xTest)
```

Download a webpage
-----------------

```julia
using Requests
url = "http://www.basketball-reference.com/boxscores/201506140GSW.html"
data = readall(get(url))
```

Extract player box scores
------------------------

```julia
using Gumbo
doc = parsehtml(data)
cleveland_table = HTMLElement{:any}
gsw_table = HTMLElement{:any}
for elem in preorder(doc.root)
    if typeof(elem) == HTMLElement{:table}
        if haskey(attrs(elem),"id")
            if attrs(elem)["id"] == "CLE_basic"
                cleveland_table = elem
            elseif attrs(elem)["id"] == "GSW_basic"
                gsw_table = elem
            end
        end
    end
end

function search_table(table)
    results = []
    for elem in preorder(table)
        if typeof(elem) == HTMLElement{:tr}
            temp_results = []
            for sub_elem in preorder(elem)
                if typeof(sub_elem) == HTMLText
                    push!(temp_results,sub_elem)
                end
            end
            push!(results,temp_results)
        end
    end
    return results
end

cleveland_results = search_table(cleveland_table)
gsw_results = search_table(gsw_table)
```
